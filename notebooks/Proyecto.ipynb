# Analisis Inicial

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargamos el archivo.
df = pd.read_csv('Informaci_n_delictiva_del_municipio_de_Bucaramanga_20241031.csv')
df
```

```python
# Informacion del dataset
df.info()
```

## Observaciones:
- El dataset tiene 100993 Filas y 26 Columnas 

# Columnas Irrelevantes
Para este proyecto, se consideran irrelevantes las siguiente columnas:
- MOVIL_VICTIMA
- MOVIL_AGRESOR
- ARTICULO
- CURSO_VIDA
- CURSO_VIDA_ORDEN
- AÑO_NUM
- MES_NUM
- RANGO_HORARIO_ORDEN
- DIA_NOMBRE_ORDEN
- LOCALIDAD
- NUM_COM
- CANTIDAD_UNICA











```python
# Eliminar las columnas irrelevantes
df = df.drop(["DESCRIPCION_CONDUCTA", "MOVIL_AGRESOR", "ARTICULO", "CURSO_VIDA", "CURSO_VIDA_ORDEN", "AÑO_NUM", "RANGO_HORARIO_ORDEN", "DIA_NOMBRE_ORDEN", "LOCALIDAD", "NUM_COM", "CANTIDAD_UNICA"], axis = 1)
df
```

# Renombramos Las columnas:
- EDAD : EDAD_AFECTADO
- DELITO_SOLO : DELITO

```python
df = df.rename(columns={
    "EDAD" : "EDAD_AFECTADO",
    "DELITO_SOLO" : "DELITO"
})
df
```

# Tipos de datos 


```python
df.dtypes
```

## Observasiones
- Cambiar EDAD_AFECTADO a int
- Cambiar DIA_NUM como int64 



```python
df['EDAD_AFECTADO'] = pd.to_numeric(df['EDAD_AFECTADO'], errors='coerce').astype('Int64')
df['DIA_NUM'] = pd.to_numeric(df['DIA_NUM'], errors='coerce')
```

```python
df.dtypes
```

### EDAD_AFECTADO

```python
filasnulas_Edad = df.isnull()["EDAD_AFECTADO"]
fn_Edad = df[filasnulas_Edad]
fn_Edad
```

```python
df_limp = df[df['EDAD_AFECTADO'].notna()].copy()
```

```python
mediana_edad = df_limp['EDAD_AFECTADO'].median()
mediana_edad
```

```python
df.loc[:, 'EDAD_AFECTADO'] = df['EDAD_AFECTADO'].fillna(mediana_edad)
```

```python
df.isnull().sum()
```

## Datos Atipicos

```python
df.boxplot(figsize=(10,4))
plt.show()
```

```python
q1 = df["EDAD_AFECTADO"].quantile(0.25)
print("Primer Cuartil: ", q1)

q3 = df["EDAD_AFECTADO"].quantile(0.75)
print("Tercer Cuartil: ", q3)

iqr = q3 - q1
print("Ragno Intercuartil: ", iqr)

lim_inf = (q1 - 1.5 * iqr)
print("Limite inferior: ", lim_inf)

lim_sup = (q3 + 1.5 * iqr)
print("Limite superior: ", lim_sup)
```

```python
fig, ax = plt.subplots(figsize=(15,4))
sns.boxplot(x="EDAD_AFECTADO", data=df)
plt.axvline(x=lim_inf, color="orange", label="Lim inferior")
plt.axvline(x=lim_sup, color="orange", label="Lim superior")
plt.legend()
plt.show()
```
